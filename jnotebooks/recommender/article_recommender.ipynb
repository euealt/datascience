```python
import pandas as pd
```

```python
articles = pd.read_csv('../../datasets/shared_articles.csv')
articles.head()
```

```python
# Filter and retain only shared contents
articles = articles[articles['eventType'] == 'CONTENT SHARED']
articles.head(3)
```

```python
articles.info()
```

```python
interactions = pd.read_csv('../../datasets/users_interactions.csv')
interactions.head(5)
```

```python
# show interaction types
interactions.eventType.value_counts()
```

```python
# Let's give weight to interaction types
interaction_weight = { 'VIEW': 1, 'LIKE': 2, 'BOOKMARK': 3, 'FOLLOW': 4, 'COMMENT CREATED': 5}
interactions['eventStrength'] = interactions.eventType.apply(lambda x: interaction_weight[x])
interactions.head()
```

```python
# Deal with cold-start. Filter out data with few interactions

interactions.groupby(['personId', 'contentId']).size()
```

```python

interactions_grp = interactions.groupby(['personId', 'contentId']).size().groupby('personId').size()
interactions_grp
```

```python
interactions_enough = interactions_grp[interactions_grp >= 5].reset_index()[['personId']]
interactions_enough.info()
```

```python
# Merge with main interaction to extract those selected ones with enough interaction data
interactions_selected = interactions.merge(
    interactions_enough,
    how='right',
    left_on = 'personId',
    right_on = 'personId'
)
interactions_selected.info()
```

```python
print(f"Number of interactions with 5 or more activities: {len(interactions_selected)}")
print(f"Number of interactions in the original dataset: {len(interactions)}")
```

```python
from sklearn.model_selection import train_test_split
interactions_train, interactions_test = train_test_split(interactions_selected, test_size=0.20, random_state=0)
interactions_train.info()
```
