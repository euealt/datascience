```python
import pandas as pd
col_names =  ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']
votes = pd.read_csv('../datasets/house-votes-84.csv', names = col_names, na_values=['?'])
```

```python
# Clean-ups
votes.dropna(inplace=True)
votes.replace(('y', 'n'), (1, 0), inplace=True)
votes.head()
```

```python
classes = votes.party.values
classes
```

```python
features = votes.drop('party', axis=1).values
features
```

```python
# Build our a Keras model with binary classification
from keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.models import Sequential


def create_model():
    model = Sequential()
    # We have 16 features - let's try with 8-unit layers
    model.add(Dense(32, input_dim=16, activation='relu'))
    model.add(Dropout(0.5))
    # Add a second hidden layer
    model.add(Dense(16, activation='relu'))
    model.add(Dropout(0.5))
    # Add a third hidden layer
    # model.add(Dense(32, activation='relu'))
    # model.add(Dropout(0.5))
    # Add a forth hidden layer
    # model.add(Dense(16, activation='relu'))
    # model.add(Dropout(0.5))
    # Output layer
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
    
    return model
```

```python
# Now Let's create our classifier
classifier = KerasClassifier(build_fn=create_model, epochs=100)
```

```python
# Now let's evaluate our success rate
from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, features, classes, cv=10)
scores.mean()
```
